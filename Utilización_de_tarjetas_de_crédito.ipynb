```python
import pandas as pd

import matplotlib.pyplot as plt

from matplotlib import colors

from matplotlib import cm

%matplotlib inline

df = pd.read_csv("trx_tarjetas_credito.csv" )



DatosY=pd.DataFrame(df.groupby(by=["City","Gender"])["Amount"].sum())

Total= DatosY.sort_values(by='Amount',ascending=True)

Total= DatosY.head(10)

plt.title("Primeras 14 Ciudades con mas gastos segun el Genero")

normdata = colors.Normalize(min(Total['Amount']), max(Total['Amount']))

colormap = cm.get_cmap("RdPu")

colores =colormap(normdata(Total['Amount']))

plt.pie(Total,labels=Total.index,autopct="%0.1f %%",colors=colores)

plt.axis("equal")

plt.show

plt.savefig("Primeras 14 Ciudades con mas gastos segun el Genero.jpg")

DatosY.to_excel('/content/Gastos_de_targetas_por.xlsx')


```

```python
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv("trx_tarjetas_credito.csv" )




DatosX=pd.DataFrame(df.groupby(by="Exp Type")["Amount"].sum())

listatipoGasto=['Bills','Entertainment','Food','Fuel','Grocery','Travel']

DatosExpType=df["Exp Type"]

DatosX

fig, ax = plt.subplots()

ax.set_title("Gastos Por Tipo",fontsize = 20)

ax.set_xlabel("Tipo de gasto",fontsize = 10)

ax.set_ylabel("Gasto",fontsize = 10)

ax.set_facecolor("darkslategray")

fig.set_facecolor("darkslategray")

ax.bar(listatipoGasto,DatosX['Amount'],linewidth=2,width=0.5, edgecolor="white")

ax.plot(listatipoGasto,DatosX['Amount'],'-.',color='red')

ax.scatter(listatipoGasto,DatosX['Amount'],color='black')

DatosX.to_excel('/content/Gastos_de_targetas_por_Tipo.xlsx')

plt.show

plt.savefig("ss.jpg")
```
